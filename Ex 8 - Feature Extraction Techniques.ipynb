# a) Scale - Invariant Feature Transform (SIFT)

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
img = cv.imread("C:/Users/asus/Downloads/kimi-no-na-wa-love-2880x1800-19857.jpg")
plt.subplot(121)
plt.imshow(img)
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
sift = cv.SIFT_create()
kp = sift.detect(gray,None)
img = cv.drawKeypoints(gray,kp,img)
plt.subplot(122)
plt.imshow(img)
plt.show()
```

# b) ORB(Oriented FAST and Rotated BRIEF)

```python
import numpy as np 
import cv2 
import matplotlib.pyplot as plt
  
query_img = cv2.imread("C:/Users/asus/Downloads/kimi-no-na-wa-love-2880x1800-19857.jpg") 
train_img = cv2.imread("C:/Users/asus/Downloads/kimi-no-na-wa-love-2880x1800-19857.jpg") 

query_img_bw = cv2.cvtColor(query_img,cv2.COLOR_BGR2GRAY) 
train_img_bw = cv2.cvtColor(train_img, cv2.COLOR_BGR2GRAY) 

orb = cv2.ORB_create() 
  
queryKeypoints, queryDescriptors = orb.detectAndCompute(query_img_bw,None) 
trainKeypoints, trainDescriptors = orb.detectAndCompute(train_img_bw,None) 
  
matcher = cv2.BFMatcher() 
matches = matcher.match(queryDescriptors,trainDescriptors) 
 
final_img = cv2.drawMatches(query_img, queryKeypoints, train_img, trainKeypoints, matches[:20],None) 
final_img = cv2.resize(final_img,(1000,650))

while(1):
    cv2.imshow("Matches",final_img)
    if cv2.waitKey(20) & 0xFF == 27:
        break
cv2.destroyAllWindows()
```
